# Plot Gradient

```python
import numpy as np
import matplotlib.pyplot as plt
A = 2
B = 10
C = 0.5
x0 = 1
twopi_by_A = 2 * np.pi / A
half_B = B / 2
sigma = 0.001 ** (0.5)
gradfactor = sigma * sigma / 2
reflect_at = 10

#x = np.array(range(0,11))
x = np.arange(0.75,10.1,0.05)
xarg = twopi_by_A * (x - x0)

eCx = np.exp(C * x)
eCx_less_one = eCx - 1.0
y = half_B / (eCx_less_one * eCx_less_one) * (twopi_by_A * eCx_less_one * np.sin(xarg) + C * eCx * np.cos(xarg))
```

```python
plt.rcParams['figure.figsize'] = [20,10]
plt.rcParams['font.size'] = 20
plt.plot(x, y)
plt.plot(9.5,0,marker='*',color='green',markersize=30)
plt.axvline(10,color='red')
plt.ylabel('gradient')
plt.xlabel('x-coordinate')
plt.text(8.1,10,'reflecting barrier ->', color='red')
plt.text(8.9,-3,'basis state', color='green')
```

# Plot Potential

```python
import numpy as np
import matplotlib.pyplot as plt
A = 2
B = 10
C = 0.5
x0 = 1
twopi_by_A = 2 * np.pi / A
half_B = B / 2
sigma = 0.001 ** (0.5)
gradfactor = sigma * sigma / 2
reflect_at = 10
```

```python
def odld_1d_potential(x, A=2, B=10, C=0.5, x0=1, normalize=True):
    twopi_by_A = 2 * np.pi / A
    half_B = B / 2

    xarg = twopi_by_A * (x - x0)

    eCx = np.exp(C * x)
    eCx_less_one = eCx - 1.0

    potential = -half_B / eCx_less_one * np.cos(xarg)

    # Needed to normalize so Minimum lies at 0
    if normalize is True:
        potential -= np.min(potential)

    return potential
```

```python
x = np.linspace(0.1, 10, 1000)
y = odld_1d_potential(x, A=A, B=B, C=C, x0=x0, normalize=False)
plt.rcParams['figure.figsize'] = [20,10]
plt.rcParams['font.size'] = 20
plt.plot(x, y)
plt.plot(9.5,odld_1d_potential(9.5, A=A, B=B, C=C, x0=x0, normalize=False),marker='*',color='green',markersize=30)
plt.axvline(10,color='red')
plt.ylabel('energy')
plt.xlabel('x-coordinate')
plt.ylim(-10, 20)
plt.text(7.9,10,'reflecting barrier ->', color='red')
plt.text(8.9,-3,'basis state', color='green')
plt.title('Absolute energy');
```

```python
x = np.linspace(0.1, 10, 1000)
y = odld_1d_potential(x, A=A, B=B, C=C, x0=x0, normalize=False)
min = np.min(y)
y -= min
plt.rcParams['figure.figsize'] = [20,10]
plt.rcParams['font.size'] = 20
plt.plot(x, y)
plt.plot(9.5,odld_1d_potential(9.5, A=A, B=B, C=C, x0=x0)-min,marker='*',color='green',markersize=30)
plt.axvline(10,color='red')
plt.ylabel('relative energy (to minimum)')
plt.xlabel('x-coordinate')
plt.ylim(-1, 20)
plt.text(7.9,10,'reflecting barrier ->', color='red')
plt.text(8.9,-3,'basis state', color='green')
plt.title('Normalized energy (energy minimum is 0)');
```
